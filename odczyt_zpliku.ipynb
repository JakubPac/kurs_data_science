```python
import numpy as np
import pandas as pd
```

```python
def fetch_financial_data(company = 'META'):
  import pandas_datareader.data as web
  return web.DataReader(name = company, data_source = 'stooq')

df = fetch_financial_data()
df.head()
```

```python
df.to_csv('FB.csv')
```

```python
df_nov = df[np.bitwise_and(df.index.month == 11, df.index.year == 2020)]
df_nov
```

```python
df_nov.to_csv('fb_nov.csv')
```
