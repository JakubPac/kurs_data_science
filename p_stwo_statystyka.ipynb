Miara rozrzutu

```python
X = [3.5, 4.0, 4.0]
Y = [2.0, 4.0, 5.0]

X_mean = np.mean(X)
Y_mean = np.mean(Y)
```

```python
b_X = 1 / len(X) * (abs(X-X_mean).sum())
b_Y = 1 / len(X) * (abs(Y-Y_mean).sum())

print(f'Sredni blad X: {b_X:.4f}')
print(f'Sredni blad Y: {b_Y:.4f}')
```

wariancja

```python
v_X = 1 / len(X) * (pow(X-X_mean, 2).sum())
v_Y = 1 / len(X) * (pow(Y-Y_mean, 2).sum())

print(f'Wariancja X: {v_X:.4f}')
print(f'Wariacja Y: {v_Y:.4f}')
```

Odchylenie standardowe

```python
o_X = np.sqrt(1 / len(X) * (pow(X-X_mean, 2).sum()))
o_Y = np.sqrt(1 / len(X) * (pow(Y-Y_mean, 2).sum()))

print(f'Odchylenie standardowe X: {o_X:.4f}')
print(f'Odchylenie standardowe Y: {o_Y:.4f}')
```

```python
print(np.std(X))
print(np.std(Y))
```

Rozk≈Çad normalny

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
sns.set()
```

```python
mu = 0
sigma = 1
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
plt.title('Rozk≈Çad nomalny - funkcja gƒôstosci')
_ = plt.plot(x, norm.pdf(x, mu, sigma))
```

```python
mu = 0
sigma = 1
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 500)
params = [(0, 1), (0, 2), (0, 0.5)]
plt.figure(figsize = (8, 8))
for mu, sigma in params:
  _ = plt.plot(x, norm.pdf(x, mu, sigma), label = f'mu={mu}, sigma={sigma}')
plt.legend()
```

```python
mu = 10
sigma = 3
x = np.linspace(mu - 8 * sigma, mu + 8 * sigma, 500)
_ = plt.plot(x, norm.pdf(x, mu, sigma), label = f'mu={mu}, sigma={sigma}')
_ = plt.plot(x, norm.pdf(x, mu - mu, sigma), label = f'mu={mu - mu}, sigma={sigma}')
_ = plt.plot(x, norm.pdf(x, mu - mu, sigma / sigma), label = f'mu={mu - mu}, sigma={sigma / sigma}')
plt.legend()
```

```python
plt.title('Rozk≈Çad dystrybuanty')
_ = plt.plot(x , norm.cdf(x, mu, sigma))
```

```python
plt.title('Rozk≈Çad normalny - funkcja prze≈ºycia')
_ = plt.plot(x , norm.sf(x, mu, sigma))
```
